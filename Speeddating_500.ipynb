```python
from sklearn import preprocessing
from scipy.io import arff
import pandas as pd
data = arff.loadarff('speeddating.arff')
df = pd.DataFrame(data[0])
df.head()
```

```python
#Convert byteStrings to Strings

str_df = df.select_dtypes([object]).stack().str.decode("utf-8").unstack()
#print(str_df)

for col in str_df:
    df[col] = str_df[col]
    
#df.head()
```

```python
df.head()
```

```python
#Count of null values in each column
df[df.columns[df.isnull().any()]].isnull().sum()
```

```python
def drop_nulls(df, num):
    cols = df[df.columns[df.isnull().any()]].isnull().sum()
    cols = pd.DataFrame(cols)
    cols.reset_index(inplace=True)
    cols.columns =['Variable','Nulls']
    cols = cols[cols.Nulls > num]
    drop_list = cols.Variable
    df = df.drop(drop_list, axis=1)
    df.dropna(inplace = True)
    df.to_csv("Speeddating_Processed.csv")
    return df
```
